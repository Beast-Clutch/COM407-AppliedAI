# Folder Setup

## Folder Structure for the Module

In this module, your work is organised using a simple, consistent folder structure.
Each week is treated as a small, self-contained AI mini-project.

```
AppliedAI/
└── week01/
    ├── notebook.ipynb
    ├── assets/
    ├── outputs/
    └── README.md
```

You will repeat this pattern for each week (`week02/`, `week03/`, …).

### What each part is for

* **`notebook.ipynb`**
  Each week the notebook we are working on can be stored in the folder for that week.

* **`assets/`**
  This folder can be used to store inputs to the model i.e. anything the model *reads*.
  Examples:

  * images
  * datasets

* **`outputs/`**
  This folder contains results produced by the model i.e. anything the model *creates*.
  Examples:

  * predictions
  * generated text
  * evaluation results

* **`README.md`**
  
  The README is an explanation of your work that week.  It is written for someone who wants to understand what you built, why you built it, and how well it works, without having to read all of your code.

  In this module, the README serves several purposes:

- **Context:** it explains the problem being addressed and who the system is for.
- **Intent:** it clarifies what the AI model is meant to do (and what it is *not* meant to do).
- **Evaluation:** it states how success is judged and summarises key observations.
- **Reflection:** it captures what worked, what didn’t, and what you would improve next.

### Why this is good practice

This structure is used because it reflects how AI projects are organised in practice:

* **Clarity** – it's always obvious what is code, what is input data, and what is output
* **Reproducibility** – someone else can rerun your work and understand what you tested
* **Good habits** – you learn to separate *experiments* from *results*
* **Scalability** – as models become more complex,your project doesn't become messy
* **Assessment-friendly** – it makes your work easier to understand and evaluate
